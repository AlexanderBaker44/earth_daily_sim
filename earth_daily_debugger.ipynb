```python

def run_trials(num_trials,const_params,scenario_params,nlt_thresh):
    total_delays = []
    sim = SimulationRun(const_params,scenario_params,nlt_thresh)
    for i in range(num_trials):
        total_delay = sim.run_sim()
        total_delays.append(total_delay)

    return np.mean(total_delays),np.std(total_delays)
        
```

```python
run_trials(10, params, scen_params,nlt_thresh)
```
